# 1.2 DATA: DICTIONARY

#### Column Descriptors:
The dataset contains the following key columns:

-Company Name: The name of the company posting the job. : The title of the job posting. <br>

-Location: Geographical location of the job. <br>

-Contract Type: Type of employment contract offered (e.g., Full-time, Part-time). <br>

-Experience Level: Required experience level for the job (e.g., Entry-level, Mid-Senior level). <br>

-Sector: Industry sector of the job posting. <br>

-Description: Brief description of the job role and responsibilities. <br>

-Applications Count: Number of applicants for the job. <br>

-At: The date when the job was posted
